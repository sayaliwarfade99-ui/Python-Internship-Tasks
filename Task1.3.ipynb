# 3. Lists in Python 

This program demonstrates different operations on Python lists such as accessing, modifying, adding, removing elements, and using various list methods. 

```python
# Creating a list
fruits = ["Apple", "Banana", "Mango", "Orange"]

print("Orignial List:", fruits)

# Accessing elements using indexing
print("Frist Element:", fruits[0])
print("Lats Elements:", fruits[-1])

# Accessing elements using slicing
print("First Two Elements:", fruits[0:2])

# Modifying an element
fruits[1] = "Pineapple"
print("After Modification:", fruits)

# Adding elements 
fruits.append("Grapes")
print("After Append:", fruits)

fruits.insert(2, "Cherry")
print("After Insert:", fruits)

#Removing elements
fruits.remove("Cherry")
print("After Remove:", fruits)

fruits.pop()
print("After Pop:", fruits)

# Other list methods
numbers = [5, 2, 8, 1, 5]

numbers.sort()
print("Sorted List:", numbers)

numbers.reverse()
print("Reversed List:", numbers)

numbers.extend([10, 29])
print("Extended List:", numbers)

print("Count of 5:", numbers.count(5))

print("Index of 8:", numbers.index(8))

copy_list = numbers.copy()
print("Copied List:", copy_list)

numbers.clear()
print("After Clear:", numbers)




```

## Explination

1. A list is used to store multiple items in a single variable.
2. Indexing is used to access individual elements from a list.
3. Slicing is used to access a range of elements.
4. Elements can be modified by assigning a new value to a specific index.
5. append() adds an element at the end of the list.
6. insert() adds an elements at a specified position.
7. remove() deletes a specified element from the list.
8. pop() removes the last elements from the list.
9. sort() arrange the elements in acending order.
10. reverse() reverses the order of elements.
11. extend() adds multiple elements to a list.
12. count() returns the number of occurrences of an element.
13. index() returns the position of an element.
14. copy() creates a duplicate copy of the list.
15. clear() removes all elements from the list.
