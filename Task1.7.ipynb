# Task 7: Common Python Errors

## Aim 
To understand common Python errors, their causes, and how to fix them.

    ## Errors Covered
    1. IndentationError
    2. NameError
    3. ValueError
    4. TypeError
    5. IndexError
    6. KeyError

```python
# Task 7: Common Python Errors

print("1. IndentationError Example")
print("Cause: Incorrect indentation in Python code.")


# Wrong Code:
# if True:
# print("Hello")

# Correct Code:
if True:
    print("Hello")

print("\n-----------------------------")


print("2. NameError Example")
print("Cause: Using a variable before defining it.")

try:
    print(student_name)
except NameError as e:
    print("Error:", e)

student_name = "Sayali"
print("Correct Output:", student_name)

print("\n-----------------------------")

print("4. TypeError Example")
print("Cause: Performing operations on incompatible data types.")

try:
    result = "10" + 5
except TypeError as e:
    print("Error:", e)

print("Correct Output:", int("10") + 5)

print("\n-----------------------------")

print("5. IndexError Example")
print("Cause: Accessing an index that does not exist.")

numbers = [10, 20, 30]

try:
    print(numbers[5])
except IndexError as e:
    print("Error:", e)

print("6. KeyError Example")
print("Cause Accessing a dictionary key that does not exist.")

student = {
    "name": "Sayali",
    "age": 23
}

try:
    print(student["city"])
except KeyError as e:
    print("Error:", e)

print("Using get():", student.get("city"))
print("Using get() with default value:", student.get("city", "Key Not Found"))




```

# Explanation

1. **IndentationError**
   - Occurs when the indentation is incorrect.
   - Python uses indentation to define blocks of code.
   - Fix: Use proper indentation

2. **NameError**
  - Occurs  when  a variable is used before it is defined.
  - Fix: Define the variable before using it.

3. **ValueError**
  - Occurs when a function receives the correct data type but an invalid value.
  - Fix: Validate input before converting or processing it.

4. **TypeError**
  - Occurs when operations are performed on incompatible data types.
  - Fix: Convert values to compatible data types before accessing an index.

5. **IndexError**
  - Occurs when accessing an index outside the range of list or tuple.
  - Fix: Check the lenght of the sequence before accessing an index.

6. **KeyError**
  - Occurs when trying to access a dictionary key that does not exist.
  - Fix: Use the 'get()' method or check if the key exists before accessing it.
