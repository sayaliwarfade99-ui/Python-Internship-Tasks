# Program 1: Grade Calculator using IF-EliF-Else

## Explanation

This program calculates the student's grade based on the marks entered by user.

Conditions:
- Marks between than 75 -> Grade 0
- Marks between 60 and 75 -> Grade A
- Marks between 50 and 60 -> Grade B
- Marks between 35 and 50 -> Grade c
- Marks less than 35 -> Fail

This program demonstrates the use of if, elif, and else statements.

```python
p = int(input("Enter Marks: "))

if p > 75:
    print("Grade: 0")
elif p > 60:
    print("Grade: A")
elif p < 50:
    print("Grade: B")
elif p >= 35:
    print("Grade: C")
else:
    print("Fail")
    
```

# Program 2: Check Divisibility

## Explanation

This program checks whether a number is divisible by 2, divisible by 3, divisible by both, or divisible by neither using if-elif-else statements.

```python
n = int(input("Enter Number: "))

if n % 2 == 0 and n % 3 == 0:
    print("Divisible by 2 and 3")
elif n % 2 == 0:
    print("Divisible by 2, not 3")
elif n % 3 == 0:
    print("Divisible by 3, not 2")
else:
    print("Not divisible by 2 and 3")
```
