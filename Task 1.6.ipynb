# 6. Dictionaries in Python

## Introduction
A dictionary is a built-in data type in Python that stores data in key-value pairs.
Each key is unique, and dictionaries are mutable, meaning they can be modified after creation.

```python
# Creating a dictionary
student = {
    "name": "Sayali",
    "age": 23,
    "course": "BCA"
}

print("Original Dictionary:", student)
```

## Step 1 : Accessing Key-Value Pairs

Dictionary values can be accessed using keys or the get() method.

```python
print("Name:", student["name"])
print("Course:", student.get("course"))
```

## Step 2: Adding an Item

A new key-value pair can be added by assigning a value to a new key

```python
student["city"] = "Nagpur"
```

## Step 3: Removing an Item

The pop() method removes a specified key and returns  its value.

```python
student.pop("age")

print(student)
```

## Step 4: Using get() Method

The get() method returns the value of the specified key.

## Step 5: Using keys() Method

The keys() method returns all keys present in the dictionary.

```python
print(student.keys())
```

## Step 6: Using values() Method

The values() method returns all values present in the dictionary.

```python
print(student.values())
```

## Step 7: Using items() Method

The items() method returns all key-value pairs in the dictionary.

```python
print(student.items())
```

## Step 8: Using update() Method

The update()  method adds new key-value pairs or updates existing values.

```python
student.update({"age": 23, "college":  "Manipal University"})

print(student)
```

## Conclusion 
In this task, we learned dictionary operations such as accessing values, adding and removing items, and using methods like get(), key(), value(), items(), pop(), and update().
