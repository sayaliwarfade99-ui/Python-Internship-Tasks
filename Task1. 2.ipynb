# Operators in Python

This program demonstrates different types of operations in python, including Arithmetic, Comprison, Logical, and Assignment operators.

- Arithmetic operators are used for mathematical calculations.
- Comparison operators are used to compare values and return True or False.
- Logical operators are used to combine conditions.
- Assigment operations are used to assign and update variable values.

The code below shows examples of each type of operator.

```python
# Arithmetic Operators
a = 20 
b = 5

print("Addition:", a + b)
print("Subtration:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)


# Comparison Operators
x = 10
y = 20

print("Equal to:", x == y)
print("Not Equal to:", x != y)
print("Less than:", x < y)
print("Greater than:", x > y)


# Logical Operators 
p = True 
q = False

print("AND Operation:", p and q)
print("OR Operation:", p or q)


# Assignment Operators 
num = 10 

num += 5
print("After += :", num)

num -= 3
print("After -= :", num)
```
