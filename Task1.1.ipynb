## Numbers in Python

```python
# Integer
a = 10

# Float 
b = 5.5 

# Complex Number
c = 2 + 3j

print("Integer:", a)
print("Float:", b)
print("Complex:", c)
```

```python
x = 20 
y = 5

print("Addition:", x + y)
print("Subtration:", x - y)
print("Multiplication:", x * y)
print("Division:", x / y)
```

### Explanation 

- Integer : Whole number such as 10, 20, -5.
- Float : Decimal number such as 5.5, 3.14.
- Complex Number : Numbers with real and imginary part such as 2+3j.

Arithmetic operations:
- Addition (+)
- Subtraction (-)
- Multiplication (*)
- Division (/)
