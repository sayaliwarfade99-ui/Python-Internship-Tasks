# 4. Tuples in Python

## Aim
To understand tuples in Python and perform basic operations such as accessing elements, slicing, concatenation, repetition, count(), and index() methods.

## Explination

A tuple is an ordered collection of elements in Python. Tuples are similar to lists, but they cannot be changed after they are created. Because of this, tuples are called immutable. They can store different types of data like numbers, strings, and boolean values.

In this program, we will learn how to access tuple elements, extract a portion of a tuple using slicing, combine tuples, repeat tuples, count the occurrences of an element, and find the position of an element.



```python
# Creating a tuple 

fruits = ("Apple", "Banana", "Mango", "Orange", "Grapes")

print("Tuple:", fruits)
```

```python
# Accessing element using indexing 

print("First element:", fruits[0])
print("Third element:", fruits[2])
print("Last element:", fruits[-1])
```

### Explanation

Indexing is used to access a specific element from a tuple.

- Index starts from 0.
- Positive indexing starts from the begining.
- Negative indexing starts from the end.

```python
# Slicing a tuple

print("Elements from index 1 to 3:", fruits[1:4])
print("First three elements:", fruits[:3])
print("Last two elements:", fruits[-2:])
```

### Explanation

Slicing is used to extract a part of a tuple.

Syntax:
tuple[start : end]

The Starting index is included, but the ending index is excluded.

```python
# Concatenating two tuples

tuple1 = (1, 2, 3)
tuple2 = (4, 5, 6)

result = tuple1 + tuple2

print("Combined Tuple:", result)
```

### Explanation

The (+) operator is used to combine two tuples into single tuple.

```python
# Repeated a tuple

numbers = (10, 20)

print("Repeated Tuple:", numbers * 3)
```

### EXplanation

The (*) operator is used to repeat the elements of a tuple multiple times.

```python
# Using count()

marks = (80, 90, 75, 90, 60, 90)

print("Number of times 90 appears:", marks.count(90))
```

### Explanation

The count () method returns how many times a specific element appears in the tuple.

```python
# Using index()

subjects = ("Python", "Java", "C", "HTML")

print("Index of Java:", subjects.index("Java"))
```

### Explanation

The index() method returns the position of the first occurrence of an element in the tuple.
