# 5. Sets in Python

## Aim
To understand Python sets and perform operations such as adding elements, removing elements, union, intersection, and difference using built-in methods.

## Explanation

A set is an unordered collection of unique elemnts in Python. It does not allow duplicate values. Sets are useful for performing mathematical operations like union, intersection, and difference.

### Methods Used

1. **add()** - Adds a new elemnt to set.
2. **remove** - Removes the specified element from the set.
3. **union()** - Combines elements from two sets.
4. **intersection()** - Returns common elements from both sets.
5. **difference()** - Returns elements that are present in the first set but not in the second.



```python
# Creating two sets

set1 = {10, 20, 30, 40}
set2 = {30, 40, 50, 60}

print("Set 1:", set1)
print("Set 2:", set2)

# Adding an element
set1.add(70)
print("\nAfter add():", set1)

# Removing an element
set1.remove(20)
print("After remove():", set1)

# Union 
union_set = set1.union(set2)
print("\nUnion:", union_set)

# Intersection
intersection_set = set1.intersection(set2)
print("Intersection:", intersection_set)

# Difference
difference_set = set1.difference(set2)
print("Difference:", difference_set)
```

## Conclusion

In this program, we learned how to work with 
Pyhton sets. We performed operations like adding and removing elements using add() and remove(). We also used union(), intersection(), and difference() method to combine and compare two sets. These operations are useful for handing unique data efficiently.
